```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
print(sns.get_dataset_names())
```

```python
crash_df=sns.load_dataset('car_crashes')
crash_df
```

```python
crash_df.describe()
```

```python
crash_df.info()
```

**DISTRIBUTION PLOT**

```python
sns.distplot(crash_df['not_distracted'])   #shift +tab for more information
```

```python
sns.distplot(crash_df['not_distracted'],kde=False)
```

```python
sns.distplot(crash_df['not_distracted'],kde=False,bins=25)
```

**joint plot**

```python
crash_df.describe()
```

```python
crash_df.info()
```

```python
sns.jointplot(x='speeding',y='alcohol',data=crash_df)
```

```python
sns.jointplot(x='speeding',y='alcohol',data=crash_df,kind='hex')   #kind= scatter or reg or kde or hex
```

**KDE PLOT**

```python
sns.kdeplot(crash_df['alcohol'])  # alcohol usgae 
```

**PAIR PLOT**--> relationship across the entire dataframe numerical values

```python
crash_df.info()
```

```python
crash_df.describe()
```

```python
sns.pairplot(crash_df)
```

```python
tips_df=sns.load_dataset('tips')
tips_df
```

```python
sns.pairplot(tips_df,hue='sex',palette='Blues')
```

**RUG PLOT**--> single column of data points in a dataframe as sticks across the bottom on this axis

```python
sns.rugplot(tips_df['tip'])
```

STYLING

```python
sns.set_style('ticks')  #white or whitegrid or dark or darkgrid or ticks
plt.figure(figsize=(8,4))
sns.set_context('paper',font_scale=2)  #paper or talk or poster
sns.jointplot(x='speeding',y='alcohol',data=crash_df,kind='reg')
#sns.despine(left=True,bottom=True)  #left-rgiht-bottom-True or False
```

**BAR PLOTS--->**

```python
tips_df.info()
```

```python
sns.barplot(x='sex',y='total_bill',data=tips_df,estimator=np.median)
```

Count plot

```python
sns.countplot(x='sex',data=tips_df)
```

box plot

```python
sns.boxplot(x='day',y='total_bill',data=tips_df,hue='sex')
plt.legend(loc=1)
```

violin plot---> box plot and kde plot combination

```python
sns.violinplot(x='day',y='total_bill',data=tips_df,hue='sex',split=True)
```

STRIP PLOT-->FIG SIZE

```python
plt.figure(figsize=(8,5))
sns.stripplot(x='day',y='total_bill',data=tips_df,jitter=True,hue='sex',dodge=True)
```
